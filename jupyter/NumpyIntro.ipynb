### Numpy Overview
Numpy is numeric python - very useful for data anaylsis

```python
# numpy needs to be imported
import numpy as np
```

```python
# Numpy offers additional data structures
# the arange is an array-range object
n = np.arange(2,8,0.5) # start, stop-before, step
n # all members must be the same data type
n.dtype # numpy will choose the relevant dtype: int or float
```

```python
# we will build an array of random values
m = np.random.randn(9,3) # we will have an array shaped 9 by 3
m.dtype
m.shape
m
```

```python
# multi-dimensional arrays
e = np.random.randn(2,3,4,5,2,3)
e # we have a six dimensional array
```

```python
# creating arrays
# careful 'empty' may choose memory that has left-over data in it!!!!!!!
f = np.empty( (3,6) ) # find some memory and create an array there
f= np.zeros( (3,6), 'int32' )
f.dtype
f
```

### Working with Arrays

```python
s = f.astype(np.str_)
s
i = s.astype(np.float32)
i
```

```python
# We may carry out calculations across arrays
z = np.zeros((3,3,3,3)) # this is a tuple of dimensions
z = np.random.randn(9000,9000)
powers = (z+3)**2
powers
```

```python
# we may apply any mathematics across all membrs of the array e.g. (z-2)*0.5/f
# we may apply slicing to arrays just like in Python
z[0:5:2]
z[::-1] # reverse the array
```

```python
# We may access any member of an array
z[0,1]=99 # mutate the value at this location
z[0,1]
```
