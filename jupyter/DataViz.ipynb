## Presenting Graphical Data

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# create a structure of arbitrary data to be charted
x = np.arange(5)
y = np.random.rand(5)+2 # all the random numbers greater than 2
(x, y)
```

```python
# we can use these values to plot a bar chart
plt.bar(x,y)
# plt.barh(x,y)
# plt.pie(y)
plt.title('Verticle Bars')
```

### Plotting Line Chart

```python
x = np.linspace(0,6,11)
y = np.power(x, 2) # this is performant
(x,y)
```

```python
# line charts are the defailt for plt
plt.axis([0,7,0,42])
plt.xlabel('Linear Progression')
plt.ylabel('Squares')
plt.title('Line chart')
plt.plot(x,y)
```
